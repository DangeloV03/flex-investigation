# Temporary: homo_eps14_Ly8 φ/ψ zoom exploration

Zoomed view of the same data as `plots/homo_eps14_Ly8_phi_psi.png`,
filtered to μ ∈ [-2.825, -2.2775]. Large canvas + tiny markers so overlapping
points near coexistence are easier to separate.

```python
import os
import numpy as np
import matplotlib
matplotlib.use("Agg")  # safe for notebook execution without display
import matplotlib.pyplot as plt
from analyzer import discover_combo_results, build_curves

RESULTS_DIR = "results"
MU_MIN, MU_MAX = -2.825, -2.2775
TAG = "homo_eps14_Ly8"
MU_COEX_SIM = -2.801232  # from manage.csv

grouped = discover_combo_results(RESULTS_DIR)
combo = None
for key, data in grouped.items():
    job = data["job"]
    if job["scheme"] == "homo" and float(job["Ly"]) == 8 and float(job["epsilon"]) == -1.4:
        combo = data
        break
if combo is None:
    raise RuntimeError("homo eps14 Ly8 combo not found in results/")

mu, phi, phi_err, psi, psi_err = build_curves(combo["points"])
mask = (mu >= MU_MIN) & (mu <= MU_MAX)
mu, phi, phi_err, psi, psi_err = mu[mask], phi[mask], phi_err[mask], psi[mask], psi_err[mask]
print(f"{len(mu)} points in μ ∈ [{MU_MIN}, {MU_MAX}]")
print(f"μ span: {mu.min():.6f} … {mu.max():.6f}")
print(f"φ span: {phi.min():.6f} … {phi.max():.6f}")
print(f"ψ span: {psi.min():.6f} … {psi.max():.6f}")
```

```python
# Tabular view — sort by μ to see ordering in the dense cluster
import pandas as pd
df = pd.DataFrame({
    "mu": mu,
    "phi": phi,
    "phi_err": phi_err,
    "psi": psi,
    "psi_err": psi_err,
}).sort_values("mu")
df
```

```python
# Zoom plot: huge figure, tiny markers, no connecting line in dense region
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(28, 12), dpi=150)

marker_kw = {"s": 12, "linewidths": 0.4, "edgecolors": "black", "zorder": 3}
err_kw = {"elinewidth": 0.6, "capsize": 1.5, "capthick": 0.4}

ax1.errorbar(mu, phi, yerr=phi_err, fmt="none", **err_kw, color="C0", zorder=1)
ax1.scatter(mu, phi, c="C0", label=r"$\phi$", **marker_kw)
ax1.axhline(0, color="gray", linestyle="--", linewidth=0.8, zorder=0)
ax1.axvline(MU_COEX_SIM, color="red", linestyle="--", linewidth=1,
            label=f"$\mu_{{coex}}^{{SIM}}={MU_COEX_SIM:.4f}$")
ax1.set_xlim(MU_MIN, MU_MAX)
phi_pad = max(0.02, (phi.max() - phi.min()) * 0.15)
ax1.set_ylim(phi.min() - phi_pad, phi.max() + phi_pad)
ax1.set_xlabel(r"$\mu$", fontsize=14)
ax1.set_ylabel(r"$\phi$", fontsize=14)
ax1.set_title(f"$\phi$ vs $\mu$ — {TAG} (zoom)", fontsize=13)
ax1.legend(fontsize=11)
ax1.grid(True, linestyle="--", alpha=0.35)
ax1.tick_params(labelsize=10)

ax2.errorbar(mu, psi, yerr=psi_err, fmt="none", **err_kw, color="C1", zorder=1)
ax2.scatter(mu, psi, c="C1", label=r"$\psi$", **marker_kw)
ax2.axvline(MU_COEX_SIM, color="red", linestyle="--", linewidth=1,
            label=f"$\mu_{{coex}}^{{SIM}}={MU_COEX_SIM:.4f}$")
ax2.set_xlim(MU_MIN, MU_MAX)
psi_pad = max(0.02, (psi.max() - psi.min()) * 0.15)
ax2.set_ylim(max(0, psi.min() - psi_pad), psi.max() + psi_pad)
ax2.set_xlabel(r"$\mu$", fontsize=14)
ax2.set_ylabel(r"$\psi$", fontsize=14)
ax2.set_title(f"$\psi$ vs $\mu$ — {TAG} (zoom)", fontsize=13)
ax2.legend(fontsize=11)
ax2.grid(True, linestyle="--", alpha=0.35)
ax2.tick_params(labelsize=10)

fig.tight_layout()
out = "plots/homo_eps14_Ly8_phi_psi_zoom_temp.png"
os.makedirs("plots", exist_ok=True)
fig.savefig(out, dpi=200, bbox_inches="tight")
plt.show()
print(f"Saved: {out}")
```

```python
# Extra-tight zoom on the dense coexistence cluster only (μ ≈ -2.834 … -2.788)
cluster_mask = (mu >= -2.834) & (mu <= -2.788)
cmu, cphi, cpe, cpsi, cspe = mu[cluster_mask], phi[cluster_mask], phi_err[cluster_mask], psi[cluster_mask], psi_err[cluster_mask]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(32, 14), dpi=150)
tiny = {"s": 20, "linewidths": 0.5, "edgecolors": "black", "zorder": 3}

for ax, y, yerr, color, ylab in [(ax1, cphi, cpe, "C0", r"$\phi$"), (ax2, cpsi, cspe, "C1", r"$\psi$")]:
    ax.errorbar(cmu, y, yerr=yerr, fmt="none", elinewidth=0.7, capsize=2, color=color, zorder=1)
    ax.scatter(cmu, y, c=color, **tiny)
    ax.axvline(MU_COEX_SIM, color="red", linestyle="--", linewidth=1)
    if ylab == r"$\phi$":
        ax.axhline(0, color="gray", linestyle="--", linewidth=0.8)
    ax.set_xlim(-2.834, -2.788)
    pad = max(0.01, (y.max() - y.min()) * 0.2)
    ax.set_ylim(y.min() - pad, y.max() + pad)
    ax.set_xlabel(r"$\mu$", fontsize=14)
    ax.set_ylabel(ylab, fontsize=14)
    ax.set_title(f"{ylab} cluster detail — {len(cmu)} pts", fontsize=13)
    ax.grid(True, linestyle="--", alpha=0.35)
    # label each point with index (sorted by μ)
    order = np.argsort(cmu)
    for i, idx in enumerate(order):
        ax.annotate(str(i), (cmu[idx], y[idx]), textcoords="offset points", xytext=(3, 3),
                    fontsize=7, color="0.3")

fig.tight_layout()
out2 = "plots/homo_eps14_Ly8_phi_psi_cluster_temp.png"
fig.savefig(out2, dpi=200, bbox_inches="tight")
plt.show()
print(f"Saved: {out2}")
```
